```python
def prob_calc_m(index_no, array_to_update):
    x = (0.88-array_to_update[index_no])/5
    for i in range(12):
        array_to_update[i] = array_to_update[i] - (x/22)
    array_to_update[index_no] = array_to_update[index_no]+(x/22)+x
    array_to_update[12] = array_to_update[12] - (x/2)
    return(array_to_update)

def prob_calc_null(remain,array_to_update):
    #remain =(window/90) *((190*90) - rows_passed)/90
    x = (0.88-array_to_update[12])/remain
    for i in range(12):
        array_to_update[i] = array_to_update[i] - x/12
    array_to_update[12] = array_to_update[12]+x
    return(array_to_update)
```
